```python
from PIL import Image
import numpy as np
import os
import cv2
```

```python
#turn from binray to 255
path = "C:\\Users\\trand\\Zebrafish_research\\MedSAM\\assets\\zebrafish_segmentation_mask_binary_testing\\"
new_path = "C:\\Users\\trand\\Zebrafish_research\\MedSAM\\assets\\zebrafish_segmentation_test_255\\"
files_list = os.listdir(path)
for file_name in files_list:
    # Load the image
    image_path = os.path.join(path, file_name)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Scale the pixel values to the range [0, 255]
    scaled_image = image * 255

    # Save the scaled image back to the same location
    new_img_path = os.path.join(new_path, file_name)
    cv2.imwrite(new_img_path, scaled_image)
```

```python
def calculate_iou(groundtruth_mask, prediction_mask):
    # Calculate intersection, union, and IoU
    intersection = np.logical_and(groundtruth_mask, prediction_mask)
    union = np.logical_or(groundtruth_mask, prediction_mask)
    iou = np.sum(intersection) / np.sum(union)

    return iou
```

```python
path = "C:/Users/trand/Zebrafish_research/zebrafish_echo_original/ground_truth_label"
list_path_groundtruth = os.listdir(path)
iou_list = []
path2 = "C:/Users/trand/Zebrafish_research/mask_sam_train_predict_original_mobilesam"
list_path_prediction = os.listdir(path2)
print(list_path_prediction)
for i in range(len(list_path_prediction)):
    img_name = list_path_prediction[i][4:-4]
    #print(img_name)
    groundtruth_img = Image.open((os.path.join(path, img_name))).convert("L")
    prediction_img = Image.open((os.path.join(path2,list_path_prediction[i]))).convert("L")
    prediction_img = prediction_img.resize(groundtruth_img.size, Image.LANCZOS)
    #turn to np array
    groundtruth_img_array = np.array(groundtruth_img)
    prediction_img_array = np.array(prediction_img)

    # # Binarize the images (optional, depending on your masks)
    # groundtruth_mask = (groundtruth_img_array > 0)
    # prediction_mask = (prediction_img_array > 0)

    iou = calculate_iou(groundtruth_img_array, prediction_img_array)
    iou_list.append(iou)
    print(f"IoU for {img_name}: {iou}")
```
